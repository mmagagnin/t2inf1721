## Criação do grafo fake

As células abaixo criam um grafo aleatório a partir de uma dada sequência de números.

```python
possible_configs = list(itertools.permutations([0,1,2,3,4,5,6,7,8]))

qtd_configs = len(possible_configs)

possible_config_names = []

for i in range(1,qtd_configs+1):
    possible_config_names.append("cfg"+str(i))
```

```python
grafo = {}
for cfg in possible_config_names:
    m_cfg = random.randint(0,100)#qtd_configs-1)
    grafo[cfg] = random.sample(possible_config_names,m_cfg)

# grafo
```

```python
parents = dict(keys=grafo.keys())
visited = dict(keys=grafo.keys())

for key in grafo.keys():
    parents[key] = None
    visited[key] = False
```

```python
def BFS_node(G,s):
    global visited
    global parents
    
    #Initialize vector os leve 0: L[0] = {}
    L = {}
    L[0] = []
    L[0].append(s)
    
    # print("L[0] = ",L[0])
    
    #Mark s as visited
    visited[s] = True
    
    #Não sei exatamente o ponto de parada do for
    for i in range(1,len(G.keys())):
        L[i] = []
        #For each u in L[i-1]
        for u in L[i-1]:
            #For each v in Adj[u]
            for v in G[u]:
                if visited[v] == False:
                    L[i].append(v)
                    parents[v] = u
                    visited[v] = True
                    
        # print("L[%d] = "%i,L[i])
        
        if len(L[i]) == 0:
            return L
```

```python
def BFS(G):
    global visited
    global parents
    
    for key in G.keys():
        visited[key] = False
        parents[key] = 0
    
    count_component = 0
    components = []
    
    for s in G.keys():
        if visited[s] == False:
            print("Launch BFS")
            components.append(BFS_node(G,s))
            count_component+=1
    
    print("Graph with %d convex components"%count_component)
    return components
             
```

```python
components = BFS(grafo)
```
